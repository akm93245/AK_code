What is data wrangling?

Data wrangling is the process of converting raw data into a usable form.\
It may also be called data mugging or data remediation.

Data wrangling describes a series of processes designed to explore, transform, and validate raw datasets from their messy and complex forms into high-quality data.\
You can use your wrangled data to produce valuable insights and guide business decisions. 

**DATA  WRANGLING STEPS**

-Discovery\
-Transformation\
-Validation\
-Publishing
