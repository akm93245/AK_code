___
### data for classification 

```python
df = sns.load_dataset('titanic')
X =  df[['pclass','sex','age','sibsp','parch','fare']]
y = df['survived']
X.sex = X.sex.map({'male': 0,'female': 1})
X.age.fillna(X.age.mean(),inplace=True)
```

___
### data for regression


```python
df = sns.load_dataset('titanic')
X =  df[['age']]
y = df['fare']
X.age.fillna(X.age.mean(),inplace=True)
```

___

# best parameter for logistic regression

```python
from sklearn.model_selection import GridSearchCV

#creating the model
model = LogisticRegression()

# define parameter grisd
grid_params = {}

# object gird search cv (creating he model)
grid_search = GridSearchCV(model , grid_params , cv=5 ,scoring='precision')

# fitting the model
grid_search.fit(X,y)

# printing the best parameter

print("Best parameter : ", grid_search.best_params_)
print("Best score : " , grid_search.best_score_)
```

```python
from sklearn.model_selection import GridSearchCV

#creating the model
model = LogisticRegression()

# define parameter grisd
grid_params = {'C':[0.1,1.0,10.0],'class_weight':['balanced']}

# object gird search cv (creating he model)
grid_search = GridSearchCV(model , grid_params ,scoring='r2')

# fitting the model
grid_search.fit(X,y)

# printing the best parameter

print("Best parameter : ", grid_search.best_params_)
print("Best score : " , grid_search.best_score_)
```

# Linear Regression

```python
from sklearn.model_selection import GridSearchCV

#creating the model
model = LinearRegression()

# define parameter grisd
grid_params = {}

# object gird search cv (creating he model)
grid_search = GridSearchCV(model , grid_params , cv=5 ,scoring='d2_absolute_error_score')

# fitting the model
grid_search.fit(X,y)

# printing the best parameter

print("Best parameter : ", grid_search.best_params_)
print("Best score : " , grid_search.best_score_)
```

```python
from sklearn.model_selection import GridSearchCV

#creating the model
model = LinearRegression()

# define parameter grisd
grid_params = {'fit_intercept':[True,False],'positive':[True,False]}

# object gird search cv (creating he model)
grid_search = GridSearchCV(model , grid_params , cv=5 ,scoring='d2_absolute_error_score')

# fitting the model
grid_search.fit(X,y)

# printing the best parameter

print("Best parameter : ", grid_search.best_params_)
print("Best score : " , grid_search.best_score_)
```
