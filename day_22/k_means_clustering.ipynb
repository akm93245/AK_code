# k-means clustering
k means will only have the X values

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= sns.load_dataset('titanic')
```

```python
df.age = df['age'].fillna(df.age.mean())
```

```python
# import requirements
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# data in features
X = df[['age','fare']]

# scaling the data
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

# model  creating
model = KMeans(n_clusters=5)
# fitting the model
model.fit(X_scaled)

# assign cluster labels to the data
df['clusters'] = model.labels_
```

```python
df.clusters.unique()
```

```python
plt.scatter(df['age'],df['fare'], c=df['clusters'] ,cmap='rainbow')
plt.show()
```

```python
plt.scatter(df['age'],df['fare'], c=df['clusters'] ,cmap='rainbow')
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],color= 'black')
plt.show()
```

```python
import plotly.express as px
```

```python
fig = px.scatter(data_frame=df,x='age',y='fare', color='clusters')
fig.show()
```

```python
fig = px.scatter_3d(data_frame=df,x='age',y='fare',z='who' ,color='clusters')
fig.show()
```
