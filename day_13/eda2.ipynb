### Analysis

**This analysis is brought to you by Ahmad**

let us analyze the man box of 0,
>The middle line between the box is median so which mean that median is 30\
>below the median another line of box(last line of box) is called quartline 1 it start form 25% of data spread\
>above the median line is Q3(quartline 3) reprsent the 75% of the data\
>the whole box represent the 50% of data spread\
>Lower T like shape represent the min value while the upper T represent the max value

finding out the outlier range 
```
IQR = inter quartile range
IQR = Q3 - Q1 * 1.5
here in this case we have 
Q3 = 40, Q1 = 23
so 40-23= 17

now:
17 * 1.5 = 25.5

> this mean that if a value is 25.5 above the box it will be an outlier and
> and if 25.5 below the box it will be an outlier
