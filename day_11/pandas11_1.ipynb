**barplot**

**scatterplot**\
remember the scatter work on numeric variable


````python
sns.scatterplot(x='age' , y='fare' ,hue='class' , data=df)
````

> **pairplot**\
> In this plot every column is ploted against another column
> column to column\
> pairplot show the comparison of the columns

````python
sns.pairplot(df)
````

**Histplot**\
>it will make the histplot of every numeric variable \
>histogram or histplot is very important plot in escpacially used in the statistic and bussinesses\
> bins :  for graph bar size\
> figsize () : figure size \
> grid: grid lines can be True or False

````python
#making of hisplot
df.hist(bins=10,figsize=(5,5),grid=False)
````

**Boxplot**\
> Box plot is an import plot for visualizing the data\
> we can determine the outliers via Boxplot \
> we can determine the standard Deviation also

````python

sns.boxplot(data=df,x='who',y='fare',hue='class')
````

**Countplot**\
countplot only accept of axis (i.e x )

````python
sns.countplot(data=df,x='survived')
````

**Boxen plot**\
It is like a father of box plot with more details init

````python

sns.boxenplot(data=df,x='who',y="fare")
````
