**Why model.compile ?**

Model compile is used to configure the model for training this method specifies how the model should be trainded by defining a speciefic components.

___

```python
# Importing libraries
import tensorflow as tf
import numpy as np
from tensorflow import keras

# Creating a Sequential model with one dense layer
model = tf.keras.Sequential([
    keras.layers.Dense(units=1, input_shape=[1])  # Single neuron, 1D input
])

# Compiling the model
model.compile(optimizer='sgd', loss='mean_squared_error')

# Defining the data
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)

# Fitting the model (training)
model.fit(xs, ys, epochs=100, verbose=0)  # verbose=0 to suppress output during training

# Making a prediction
# Prediction (input as NumPy array)
print("Prediction for x=10:", model.predict(np.array([10.0])))
```

```python
# Importing libraries
import tensorflow as tf
import numpy as np
from tensorflow import keras

# Creating a Sequential model with one dense layer
model = tf.keras.Sequential([
    keras.layers.Dense(units=1, input_shape=[1])  # Single neuron, 1D input
])

# Compiling the model
model.compile(optimizer='sgd', loss='mean_squared_error')

# Defining the data
xs = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)

# Fitting the model (training)
model.fit(xs, ys, epochs=500, verbose=0)  # verbose=0 to suppress output during training

# Making a prediction
# Prediction (input as NumPy array)
print("Prediction for x=10:", model.predict(np.array([10.0,20])))
```
