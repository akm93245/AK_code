```python
ag.loc[::-1].head()
```

reversing back the row order

```python
ag.loc[::-1].reset_index(drop=True).head()
```

reversing the columns

```python
ag.loc[:,::-1].head()
```

```python
ag.loc[:,::1].head()
```
