```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = sns.load_dataset('titanic')
```

```python
df_linear = df[['age','fare']]
df_linear.age.fillna(df_linear.age.mean(),inplace=True)
```

```python
df_linear
```

```python
df_linear.isnull().sum()
```

```python
sns.lmplot(data=df_linear,x='age',y='fare')
```

```python
df_linear = df_linear[(df_linear['age'] > 5 )&(df_linear['age']<54)]
```

```python
df_linear = df_linear[(df_linear['fare'] > 1 ) & (df_linear['fare'] <40 )]
```

```python
df_linear.shape
```

```python
X = df_linear[['age']]
Y = df_linear['fare']
```

```python
sns.lmplot(data=df_linear,x='age',y='fare')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# model call
model = LinearRegression()

#fit model 
model.fit(X,Y)
```

```python
#predict
model.predict([[80]])
```

```python
# model evaluation with matrices
#r2 (R2 score)
# mean square eror
#MAE (Mean absolute error)
#RMAE (Root Mean Squared Error)

X_train,X_test,Y_train,Y_test  = train_test_split(X,Y, train_size=0.80, random_state=0)

#model
model = LinearRegression()

#taining
model.fit(X_train,Y_train)

#prediction 
predictions = model.predict(X_test)

#comparing
sns.scatterplot(x=Y_test,y=predictions)
```

we can generally see that accuracy is not good



Now let us check the accuracy using the metrices

```python
from sklearn.metrics import mean_absolute_error ,mean_squared_error,r2_score
```

```python
#let us check the accuracy
model.score(X_test,Y_test)
```

```python
MAE = mean_absolute_error(y_true=Y_test , y_pred=predictions)
MSE = mean_squared_error(y_true=Y_test , y_pred=predictions)
R2 = r2_score(y_true=Y_test , y_pred=predictions)
RMSE = mean_squared_error(y_true=Y_test , y_pred=predictions,  squared=False)
```

```python
print(f"Mean Absolute Error : {MAE}")
print(f"Mean Squared Error {MSE}")
print(f"Root Mean Squared Error : {RMSE}")
print(f"R2 Score : {R2}")
```
